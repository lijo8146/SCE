```python
import rasterio
import matplotlib.pyplot as plt

def read_sentinel2_bands(sentinel2_path):
    bands = {}
    with rasterio.open(sentinel2_path) as dataset:
        for i in range(1, dataset.count + 1):
            band_name = dataset.descriptions[i - 1]
            band_data = dataset.read(i)
            bands[band_name] = band_data
    return bands

def extract_spectral_signature(bands_data, pixel_coords):
    spectral_signature = {}
    for band_name, band_data in bands_data.items():
        spectral_values = band_data[:, pixel_coords[0], pixel_coords[1]]
        spectral_signature[band_name] = spectral_values
    return spectral_signature

def plot_spectral_signature(spectral_signature):
    plt.figure(figsize=(10, 6))
    for band_name, spectral_values in spectral_signature.items():
        plt.plot(spectral_values, label=band_name)
    plt.xlabel("Band Index")
    plt.ylabel("Reflectance")
    plt.title("Spectral Signature")
    #plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Replace 'path_to_sentinel2_data' with the actual path to your Sentinel-2 dataset.
    sentinel2_path = r"C:\Users\lilly\Documents\Sentinel downloads\Santa Barbara\2021\Dec 21\S2B_MSIL1C_20211231T221559_N0301_R129_T60HVC_20211231T231427.SAFE\MTD_MSIL1C.xml"
    
    # Choose a pixel location (row, column) to plot its spectral signature.
    pixel_coords = (8996, 5932)  # Change to the desired pixel coordinates.

    bands_data = read_sentinel2_bands(sentinel2_path)
    spectral_signature = extract_spectral_signature(bands_data, pixel_coords)
    
    # Print spectral values before plotting
    for band_name, spectral_values in spectral_signature.items():
        print(f"{band_name}: {spectral_values}")
    
    plot_spectral_signature(spectral_signature)
```
