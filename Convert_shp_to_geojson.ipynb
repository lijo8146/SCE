```python
import ogr
import json

# Define input shapefile and output GeoJSON file paths
input_shapefile = r'C:\Users\lilly\Documents\SCE_study_area.shp'
output_geojson = r'C:\Users\lilly\Documents\SCE_zone.geojson'

# Open the input shapefile
input_ds = ogr.Open(input_shapefile)

# Create a GeoJSON driver
geojson_driver = ogr.GetDriverByName('GeoJSON')

# Create an output GeoJSON file
output_ds = geojson_driver.CreateDataSource(output_geojson)

# Loop through each layer in the input shapefile and copy it to the GeoJSON file
for i in range(input_ds.GetLayerCount()):
    input_layer = input_ds.GetLayerByIndex(i)
    output_layer = output_ds.CreateLayer(input_layer.GetName(), input_layer.GetSpatialRef(), ogr.wkbUnknown)

    for feature in input_layer:
        output_layer.CreateFeature(feature)

# Clean up and close the datasets
input_ds = None
output_ds = None

print("Conversion completed.")
```
